**Contact Directory**

```python
# Initial 2D list of contacts: [name, phone, email]
contacts = [
    ["Alice Johnson", "555-1234", "alice@example.com"],
    ["Bob Smith", "555-5678", "bob@example.com"],
    ["Charlie King", "555-8765", "charlie@example.com"]
]

# --- Add a new contact ---
name = input("Enter name: ")
phone = input("Enter phone number: ")
email = input("Enter email: ")

contacts.append([name, phone, email])
print("\nContact added successfully!\n")

# --- Display all contacts sorted alphabetically ---
contacts.sort(key=lambda x: x[0])  # Sort by name

print("All Contacts (Sorted):")
for c in contacts:
    print(f"Name: {c[0]}, Phone: {c[1]}, Email: {c[2]}")

# --- Display only the first 3 contacts ---
print("\nFirst 3 Contacts:")
for c in contacts[:3]:  # slicing
    print(f"Name: {c[0]}, Phone: {c[1]}, Email: {c[2]}")

# --- Search for contacts (partial name match) ---
search_term = input("\nSearch for name: ").lower()

print("\nSearch Results:")
found = False

for c in contacts:
    if search_term in c[0].lower():
        print(f"Name: {c[0]}, Phone: {c[1]}, Email: {c[2]}")
        found = True

if not found:
    print("No matching contacts found.")
```
