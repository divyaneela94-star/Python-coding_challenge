**Student Grades & Hobbies Tracker**

```python
# --- Student Grades Dictionary ---
student_grades = {}

# Add or update grades for 3 students
for i in range(3):
    name = input(f"Enter the name of student {i+1}: ")
    grade = input(f"Enter the grade for {name}: ")
    student_grades[name] = grade

print("\nüìò Student Grades:")
print(student_grades)

# --- Hobbies Set ---
hobbies = set()

print("\nEnter at least 3 hobbies:")
for i in range(3):
    hobby = input(f"Hobby {i+1}: ")
    hobbies.add(hobby)

print("\nüé® Hobbies Set:")
print(hobbies)

# Convert to frozenset
hobby_frozenset = frozenset(hobbies)

print("\n‚ùÑÔ∏è Frozenset of Hobbies:")
print(hobby_frozenset)
\
```
