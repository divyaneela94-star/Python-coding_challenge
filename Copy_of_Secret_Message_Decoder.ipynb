**[Secret Message Decoder]**

```python
# Ask user for input
message = input("Enter your secret message: ")

# Convert to uppercase
message_upper = message.upper()

# Replace spaces with underscores
message_cleaned = message_upper.replace(" ", "_")

# Show every 3rd character
every_third = message_cleaned[::3]

# Reverse the message
reversed_msg = message_cleaned[::-1]

# Display output
print("\n--- Secret Message Report ---")
print("Original Length:", len(message))
print("Uppercase Message:", message_upper)
print("Cleaned Message:", message_cleaned)
print("Every 3rd Character:", every_third)
print("Reversed Message:", reversed_msg)
```
