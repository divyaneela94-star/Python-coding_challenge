# **Step 1: Import NumPy**

```python
import numpy as np
```

**Step 2: Create 1D, 2D, and 3D Arrays**


```python
#1D Array

arr1d = np.array([10, 20, 30, 40, 50])
```

```python
# 2D Array

arr2d = np.array([[1, 2, 3],
                  [4, 5, 6]])
```

```python
# 3d Array

arr3d = np.array([[[1, 2], [3, 4]],
                  [[5, 6], [7, 8]]])
```

**Step 3: Arithmetic Operations (1D Arrays)**

```python
a = np.array([10, 20, 30])
b = np.array([5, 10, 15])

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```

**ðŸ”¹ Step 4: NumPy Functions**

*âœ… Mean, Sum, Standard Deviation*

```python
print("Mean:", np.mean(a))
print("Sum:", np.sum(a))
print("Standard Deviation:", np.std(a))
```

**ðŸ”¹ Step 5: reshape()**

```python
reshaped_array = np.array([1, 2, 3, 4, 5, 6]).reshape(2, 3)
print(reshaped_array)
```

**ðŸ”¹ Step 6: arange() and linspace()**

```python
# arange
arr_range = np.arange(0, 10, 2)
print("arange:", arr_range)

# linspace
arr_line = np.linspace(1, 10, 5)
print("linspace:", arr_line)
```

**ðŸ”¹ Step 7: Find Index of Max & Min**

```python
array = np.array([12, 45, 7, 89, 23])

print("Max Value Index:", np.argmax(array))
print("Min Value Index:", np.argmin(array))
```
