**Retail Inventory Manager**

```python
# Fixed categories (tuple - immutable)
product_categories = ("Electronics", "Stationery", "Clothing")

# Dynamic list of products
products = ["Laptop", "Pen", "T-Shirt"]

print("------ RETAIL INVENTORY MANAGER ------")
print("Product Categories:", product_categories)
print("Current Products:", products)

# Add new product
new_product = input("\nEnter a new product to add: ")
products.append(new_product)

# Remove product by index
print("\nUpdated Product List:", products)
remove_index = int(input("Enter the index of the product to remove (0 to {}): ".format(len(products)-1)))

# Ensure valid index
if 0 <= remove_index < len(products):
    removed_item = products.pop(remove_index)
    print(f"Removed: {removed_item}")
else:
    print("Invalid index. No item removed.")

# Final output
print("\n------ FINAL INVENTORY STATUS ------")
print("Categories:", product_categories)
print("Products:", products)
```
