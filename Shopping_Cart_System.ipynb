Shopping Cart System

```python
# Shopping Cart System

subtotal = 0
item_count = 0

while True:
    try:
        price = float(input("Enter item price (0 to finish): "))

        # Sentinel value to stop
        if price == 0:
            break

        # Reject negative values
        if price < 0:
            print("Negative prices are not allowed. Please enter again.")
            continue

        # Add to totals
        subtotal += price
        item_count += 1

    except ValueError:
        print("Invalid input. Please enter a valid number.")

# Determine discount rate
if subtotal >= 1000:
    discount_rate = 0.20
elif subtotal >= 500:
    discount_rate = 0.10
elif subtotal >= 200:
    discount_rate = 0.05
else:
    discount_rate = 0.00

discount_amount = subtotal * discount_rate
final_total = subtotal - discount_amount

# Bill Summary
print("\n----- BILL SUMMARY -----")
print(f"Total items: {item_count}")
print(f"Subtotal: ${subtotal:.2f}")
print(f"Discount applied: {discount_rate * 100:.0f}%")
print(f"Discount amount: ${discount_amount:.2f}")
print(f"Final total: ${final_total:.2f}")
print("------------------------")
.
```
